```python
# import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# read the data 
data = pd.read_csv('/Users/chirath/Python/IIT/Waste_Management_System-DSGP-/ALL/Dataset22-24_cleaned_model.csv')
```

```python
data.head()
```

```python
# convert the Dump Date column to datetime
data['Dump Date'] = pd.to_datetime(data['Dump Date'])
```

```python
# visualize the msw data of the year 2022 by the week
data_2022 = data[data['Dump Date'].dt.year == 2022]
data_2022['Week'] = data_2022['Dump Date'].dt.week
data_2022 = data_2022.groupby('Week').sum()
data_2022['MSW Wastage Amount (Kg)'].plot(kind='bar', figsize=(15, 6))
plt.title('MSW of the year 2022 by the week')
plt.xlabel('Week')
plt.ylabel('MSW')
plt.show()
```

```python
# get the total MSW and give me a percentage of the total MSW for each week
data_2022['Percentage'] = data_2022['MSW Wastage Amount (Kg)'] / data_2022['MSW Wastage Amount (Kg)'].sum() * 100
data_2022['Percentage'].plot(kind='bar', figsize=(15, 6))
plt.title('Percentage of MSW of the year 2022 by the week')
plt.xlabel('Week')
plt.ylabel('Percentage')
plt.show()
```

```python
# which weeks has the highest MSW top 10
data_2022['Percentage'].nlargest(15).plot(kind='bar', figsize=(15, 6))
plt.title('Top 10 weeks with the highest MSW of the year 2022')
plt.xlabel('Week')
plt.ylabel('Percentage')
plt.show()
```

```python
# visualize the msw data of the year 2022 by week number of routes (01 - VILAWALA, 29B - BORUPANA)
data_2022 = data[data['Dump Date'].dt.year == 2022]
data_2022['Week'] = data_2022['Dump Date'].dt.week
data_2022 = data_2022.groupby(['Week', 'Route']).sum()
data_2022['MSW Wastage Amount (Kg)'].unstack().plot(kind='bar', figsize=(15, 6))
plt.title('MSW of the year 2022 by the week')
plt.xlabel('Week')
plt.ylabel('MSW')
plt.show()




```

```python
# what are the top 5 routes that generate the most msw in the year 2022 in week 1
data_2022_week1 = data[(data['Dump Date'].dt.year == 2022) & (data['Month'] == 12)]
top_10_routes = data_2022_week1.groupby('Route').sum().sort_values(by='MSW Wastage Amount (Kg)', ascending=False).head(30)
top_10_routes['MSW Wastage Amount (Kg)'].plot(kind='bar', figsize=(15, 6))
plt.title('Top 10 routes that generate the most MSW in the year 2022 in week 1')
plt.xlabel('Route')
plt.ylabel('MSW')
plt.show()
```

```python
# visualize the msw data of the year 2022 by week number of routes in separate plots
data_2022 = data[data['Dump Date'].dt.year == 2022]
data_2022['Week'] = data_2022['Dump Date'].dt.week
data_2022 = data_2022.groupby(['Week', 'Route']).sum()
data_2022 = data_2022.unstack()
data_2022['MSW Wastage Amount (Kg)'].plot(kind='bar', subplots=True, figsize=(15, 100))
plt.show()
```

```python
data_2022 = data[data['Dump Date'].dt.year == 2022]
unique_routes = data_2022['Route'].unique()

for route in unique_routes:
    route_data = data_2022[data_2022['Route'] == route]
    route_data['Week'] = route_data['Dump Date'].dt.week
    plt.figure(figsize=(10, 6))
    sns.boxplot(x='Week', y='MSW Wastage Amount (Kg)', data=route_data)
    plt.title(f'Boxplot of MSW of the year 2022 for route {route}')
    plt.show()
```

```python
# identify the top 10 instances where the MSW is the highest in the year 2022 of Route == 01 - VILAWALA
data_2022_route_01 = data[(data['Dump Date'].dt.year == 2022) & (data['Route'] == '01 - VILAWALA')]
data_2022_route_01 = data_2022_route_01.groupby('Dump Date').sum()
data_2022_route_01['MSW Wastage Amount (Kg)'].nlargest(10).plot(kind='bar', figsize=(15, 6))
plt.title('Top 10 instances where the MSW is the highest in the year 2022 of Route == 01 - VILAWALA')
plt.xlabel('Dump Date')
plt.ylabel('MSW')
plt.show()
```

```python
# visualize the msw data of the year 2023 over the time period for each route
data_2023 = data[data['Dump Date'].dt.year == 2022]
data_2023 = data_2023.groupby(['Dump Date', 'Route']).sum()
data_2023 = data_2023.unstack()
data_2023['MSW Wastage Amount (Kg)'].plot(kind='line', subplots=True, figsize=(15, 100))
# on top of this add the weekly average
data_2023_weekly = data_2023.resample('W').mean()
data_2023_weekly['MSW Wastage Amount (Kg)'].plot(kind='line', subplots=True, figsize=(15, 100))
plt.show()
    
```
