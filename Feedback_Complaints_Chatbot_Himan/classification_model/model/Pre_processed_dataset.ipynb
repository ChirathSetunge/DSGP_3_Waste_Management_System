```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_excel("Waste Management Dataset.xlsx")
print("Dataset shape:", data.shape)
print("\nClass distribution:")
print(data['Category'].value_counts())
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Category', data=data)
plt.title('Class Distribution')
plt.xticks(rotation=45)
plt.show()
```

```python
print("Missing values in dataset:")
print(data.isnull().sum())

data = data.dropna(subset=['Query', 'Category'])
print("\nDataset shape after dropping missing values:", data.shape)
```

```python
print("Number of duplicate rows:", data.duplicated().sum())
data = data.drop_duplicates()
print("Dataset shape after removing duplicates:", data.shape)
```

```python
data.to_excel('cleaned_data.xlsx', index=False)
```
